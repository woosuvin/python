# `matplotlib.pyplot` 모듈을 사용한 시각화

1. `figure` 객체 생성
2. `figure` 객체에 subplot(axis)을 생성해서 추가
3. subplot에 그래프 그림
4. 그래프의 여러가지 옵션들을 설정
5. 그래프 보여주기

```python
# 1. figure 객체 생성
figure = plt.figure()

# 2. 생성된 figure 객체에 subplot(axis)을 추가
# axis = figure.add_subplot(1, 1, 1) # row, column, index
axis = figure.add_subplot(111)

# 3. subplot에 그래프 종류를 선택해서 그림
axis.hist(x=tips['total_bill'], color='pink')

# 4. 그래프에 옵션 설정
axis.set_xlabel('Total Bill')
axis.set_ylabel('Count')
axis.set_title('Histogram')

# 5. 그래프 보여주기
plt.show()
```

> figure에 두 개의 subplot을 상하로 배치하고 그래프 그림

```python
# 1. figure 객체 생성
fig = plt.figure(figsize=(6, 8))

# 2. 생성된 figure 객체에 subplot(axis)을 추가
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(2, 1, 2)

# 3. subplot에 그래프 종류를 선택해서 그림
ax1.hist(x=tips.total_bill, color='pink')
ax2.scatter(x=tips.total_bill, y=tips.tip, color='pink')

# 4. 그래프에 옵션 설정
ax1.grid()
ax2.grid()

# 5. 그래프 보여주기
plt.show()
```

figure에 2개의 subplot을 좌우로 배치하고 hist, scatter 그래프 그리기

```python
# 1. figure 객체 생성
fig = plt.figure(figsize=(10, 4))

# 2. 생성된 figure 객체에 subplot(axis)을 추가
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

# 3. subplot에 그래프 종류를 선택해서 그림
ax1.hist(x=tips.tip, color='pink')
ax2.scatter(x=tips.total_bill, y=tips.tip, color='pink')

# 4. 그래프에 옵션 설정
ax1.grid()
ax1.set_xlabel('Tip')
ax1.set_ylabel('Count')
ax1.set_title('Histogram')

ax2.grid()
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')
ax2.set_title('Scatter Plot')

# 5. 그래프 보여주기
plt.show()
```

## Histogram
* 연속형 변수를 구간으로 나눠서, 그 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.
* 연속형 변수의 분포를 알 수 있음

> hist 함수 파라미터:
* x: x축에 사용할 변수 이름. 히스토그램을 그리기 위한 자료.
* bins: 막대 개수, 구간
* color: 막대 안쪽을 채우는 색깔
* edgecolor: 막대 테두리 색깔
* density: False - 개수, True - 밀도(비율)

```python
# 1. figure 생성 2. subplot 1개 생성 3. histogram 그림
plt.hist(x=tips.total_bill, color='yellow', edgecolor='black', bins=10, density=True)

# 4. 옵션 설정
plt.xlabel('Total Bill')
plt.ylabel('Count')
plt.title('Histogram')
plt.grid()

# 5. 보여주기
plt.show()
```

## 산점도(Scatter plot)
두 변수(컬럼) 사이의 상관 관계를 파악하기 위해서 사용.

```python
# tip ~ total_bill 상관 관계 (왼쪽:y축 ~ 오른쪽:x축)
plt.scatter(x=tips.total_bill, y=tips.tip,
            alpha=0.9, # alpha: 0(투명) ~ 1.0(불투명)
            s=150,      # s: size. 마커(점)의 크기
            c='pink',  # c: color. 마커의 색깔.
            marker='*', # marker: 마커의 종류(모양). 기본값 - 'o', 삼각형 - '^', 역삼각형 - 'v', 별 - '*', 정사각형 - 's',,,
            edgecolors='gray'
            )

plt.grid()
plt.xlabel('Total bill')
plt.ylabel('Tip')

plt.show()
```

```python
# tip ~ size 상관 관계
plt.scatter(x=tips['size'], y=tips.tip,
            alpha=0.5,
            s=150,
            c='purple',
            marker='*'
            )

plt.grid()
plt.xlabel('Size')
plt.ylabel('Tip')

plt.show()
```

## 선 그래프(Line plot)
주로 시계열 데이터(시간에 따라서 변하는 데이터: 주가, 환율, 날씨, ...)를 시각화.

```python
np.random.seed(1) # 난수 생성하는 순서 설정.
df = pd.DataFrame(data={
    'time': np.arange(1, 11), # [1, 2, 3, ... , 9, 10]
    'price': np.random.rand(10) # [0, 1.0) 범위의 난수 10개를 갖는 1차원 배열
})
df
```

```python
# price ~ time
plt.plot(df.time, df.price, 'ko--')
# format 문자열(3번째 arg): 색깔, 마커 종류, 선의 종류
# 색깔: b(lue), r(ed), g(reen), (blac)k, y(ellow), ...
# 마커 종류: ㅐ, ^, v, *, s, +, ...
# 선 종류: -(solid line), :(dotted line), --(dashed line), ...

plt.grid()
plt.xlabel('Time')
plt.ylabel('Price')

plt.show()
```

```python
plt.plot(df.time, df.price,
         c='pink',
         marker='*',
         linestyle='dotted'
         )

plt.show()
```

> 1개의 subplot에 선 그래프 3개를 중첩시켜서 표현.

```python
np.random.seed(1)
df = pd.DataFrame(data={
    'time': np.arange(1, 11),
    'price1': np.random.rand(10),
    'price2': np.random.rand(10),
    'price3': np.random.rand(10)
})
df
```

```python
# price1 ~ time
plt.plot(df.time, df.price1, 'bo-', label='price1')
plt.plot(df.time, df.price2, 'rs--', label='price2')
plt.plot(df.time, df.price3, 'g+:', label='price3')

plt.legend() # 범례(legend) 표시
plt.grid()

plt.show()
```

## Ex 1.
* 4개의 subplot을 2x2 모양으로 배치.
* 4개의 subplot에 각각 성별 빈도수, 흡연여부 빈도수, 요일 빈도수, 시간 빈도수 그래프를 그리세요.

```python
sex_count = tips.sex.value_counts()
smoker_count = tips.smoker.value_counts()
day_count = tips.day.value_counts()
time_count = tips.time.value_counts()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2)
ax[0][0].bar(x=sex_count.index, height=sex_count.values, width=0.5) # ax[0, 0]: numpy에서 2차원배열 표기법
ax[0][1].bar(x=smoker_count.index, height=smoker_count.values, width=0.5)
ax[1][0].bar(x=day_count.index, height=day_count.values, width=0.5)
ax[1][1].bar(x=time_count.index, height=time_count.values, width=0.5)

for i in range(2):
  for j in range(2):
    ax[i, j].grid()

plt.show()
```

```python
categories = ['sex', 'smoker', 'day', 'time']
counts = [tips[x].value_counts() for x in categories]

fig, ax = plt.subplots(ncols=2, nrows=2)
for i in range(2):
  for j in range(2):
    idx = i * 2 + j
    ax[i, j].bar(x=counts[idx].index, height=counts[idx].values)
    ax[i, j].grid()
    ax[i, j].set_title(categories[idx])

plt.show()
```

## Ex 3.
* `tip ~ total_bill`의 상관관계를 scatter plot으로 그리세요.
* 요일별로 점의 색깔, 모양을 다르게 표현하세요.
* 범례를 표시.

```python
days = tips.day.unique()

for i in days:
  tip_by_day = tips[tips.day == i].tip
  total_bill_by_day = tips[tips.day == i].total_bill
  plt.scatter(x=total_bill_by_day,
              y=tip_by_day,
              label=i,
              alpha=0.3
              )

plt.legend()
plt.grid()

plt.show()
```

```python
days = tips.day.unique()
markers = ['o', '^', 's', '*']
colors = ['red', 'orange', 'pink', 'yellow']

for d, m, c in zip(days, markers, colors):
  subset = tips[tips.day == d]
  plt.scatter(x=subset.total_bill, y=subset.tip, c=c, marker=m, label=d)

plt.grid()
plt.legend()
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
