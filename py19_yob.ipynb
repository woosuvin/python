```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 데이터셋 준비
* YOB: Year Of Birth
* github: https://github.com/wesm/pydata-book/tree/3rd-edition/datasets/babynames
* 1980년 ~ 2010년까지 년도별로 txt파일이 저장되어 있음.
* 각 파일에는 '이름, 성별, 신생아 수'가 comma(,)로 구분되어 있음.
* (주의) 컬럼 이름은 파일에 없음.

```python
file_path='https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/babynames/yob1880.txt'
```

```python
yob1880 = pd.read_csv(file_path)
```

```python
yob1880.head()
```

> `pd.read_cvs()` 함수는 파일의 첫번째 줄을 데이터프레임의 컬럼 이름으로 취급함.  
cvs 파일에서 컬럼 이름이 포함되어 있지 않은 경우(파일의 첫번째 줄부터 데이터가 시작되는 경우)에는 컬럼 이름들을 함수 호출할 때 `names` 파라미터에 argument로 전달해야 함.

```python
col_names = ['name', 'gender', 'births'] # 데이터프레임 컬럼 이름 리스트
```

```python
yob1880 = pd.read_csv(file_path, names=col_names)
```

```python
yob1880.head()
```

> 1880년 ~ 2010년까지 각각의 txt 파일들을 읽어서 만들어진 연도별 데이터프레임을 저장하는 리스트를 만드세요.  
(결과) 131개의 데이터프레임을 저장하는 리스트. 각각의 데이터프레임에는 year 컬럼을 추가.

```python
yobs = [] # 연도별 데이터프레임들을 저장하기 위한 리스트.
for year in range(1880, 2011): # 1880 ~ 2010년까지 반복.
  # txt 파일이 저장된 깃허브 주소
  file_path=f'https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/babynames/yob{year}.txt'

  # 깃허브의 txt 파일을 읽어서 데이터프레임 생성
  df = pd.read_csv(file_path, names=col_names)

  # 데이터프레임 year 컬럼을 추가
  df['year'] = year

  # 데이터프레임을 리스트에 추가
  yobs.append(df)
```

> 데이터프레임 확인

```python
len(yobs) # 2010 - 1880 + 1
```

```python
# 리스트의 첫번째 데이터프레임 - 1880년
yobs[0].head()
```

```python
# 리스트에서 가장 마지막 데이터프레임 -2010년
yobs[-1].head()
```

> 131개의 데이터프레임들을 1개의 데이터프레임으로 합침

```python
yob_df = pd.concat(yobs, ignore_index=True)
# ignore_index: 기본값은 false, True인 경우에는,
# 각 데이터프레임들의 기존 인덱스를 무시하고 재색인(re-indexing)을 함.

yob_df.head()
```

```python
yob_df.tail()
```

```python
yob_df.info()
```

# 연도별 신생아 수, 시각화

```python
births_by_year = yob_df.groupby('year').births.sum() # 결과: Series
births_by_year
```

```python
plt.plot(births_by_year.index, births_by_year.values)
plt.grid()
plt.xlabel('Year')
plt.ylabel('# of birth')
plt.show()
```

```python
plt.bar(births_by_year.index, births_by_year.values)
plt.grid()
plt.xlabel('Year')
plt.ylabel('# of birth')
plt.show()
```

```python
sns.lineplot(x=births_by_year.index, y=births_by_year.values)
plt.grid()
plt.show()
```

```python
sns.lineplot(data=yob_df, x='year', y='births', estimator='sum', errorbar=None)
plt.grid()
plt.show()
```

```python
sns.barplot(data=yob_df, x='year', y='births', estimator='sum', errorbar=None)
plt.show()
```

# 연도별 성별 신생아 수 , 시각화

```python
sns.catplot(data=yob_df, x='year', y='births', col='gender', estimator='sum', errorbar=None, kind='bar')
plt.show()
```

```python
births_by_year_gender = yob_df.groupby(['year', 'gender']).births.sum()
births_by_year_gender
```

```python
births_pivot = yob_df.pivot_table(values='births',
                                  index='year',
                                  columns='gender',
                                  aggfunc='sum')
births_pivot
```

```python
plt.plot(births_pivot.index, births_pivot.F, label='Female')
plt.plot(births_pivot.index, births_pivot.M, label='Male')
plt.legend()
plt.grid()
plt.show()
```

```python
df = births_by_year_gender.reset_index()
# reset_index(): Series, DataFrame의 인덱스들을 DataFrame의 컬럼으로 변환
```

```python
df
```

```python
df_male = df[df.gender == 'M']
df_female = df[df.gender == 'F']
```

```python
plt.plot(df_female.year, df_female.births, label='Female')
plt.plot(df_male.year, df_male.births, label='Male')
plt.legend()
plt.grid()
plt.show()
```

```python
sns.lineplot(data=yob_df, x='year', y='births', hue='gender', estimator='sum', errorbar=None)
plt.grid()
plt.show()
```
